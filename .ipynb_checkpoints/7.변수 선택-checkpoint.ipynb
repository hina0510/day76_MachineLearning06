### feature selection(변수 선택법)
* 필요한 변수만 선택하여 모델의 예측력을 높이거나 과적합을 줄이는 것
* 알고리즘의 정확도가 유사하다면 변수의 개수가 적을수록 속도가 빠르다
* 방법
    - L1(라쏘)을 이용하여 영향력이 0(작은)값을 빼는 방법
    - 단변량(filter) : 특정 컬럼이 또 다른 컬럼과 상관관계가 있는지 확인하는 방법
    - 전진/후진 선택법(wrapper) : 컬럼을 하나씩 추가하며 점수 확인 / 하나씩 빼며 점수 확인
    - RFE : 사이킷런에서 제공하는 라이브러리를 이용하여 변수의 중요도를 확인하여 변수를 선택하는 방법
    - embed(임베드) : 알고리즘의 성질 이용
        - 알고리즘 내부에서 각 feature 구송 요소들이 얼마나 사용 되었는지를 알아보는 방식
        - 알고리즘들을 이용해 변수의 중요도를 파악, 중요도 낮은 컬럼은 지울 수 있다
        - 참고
            * RFE와 embed 방식은 특정 알고리즘에 존재하며, 해당 방식을 사용할 수 없는 알고리즘들이 존재한다.(KNN, SVR(kernel=rbf)등 사용 못함)
            * Tree계열 알고리즘들은 사용 가능하다(Random Forest, Decision Tree 등)

### wrapper(전진/후진 선택)

```python
from itertools import combinations

sample_bag = [1,2,3,4]

for c in combinations(sample_bag, 2) : 
    print(c, type(c))
```

```python
f = ['season', 'holiday', 'workingday', 'weather', 'temp',
       'atemp', 'humidity', 'windspeed', 'year', 'month', 'day', 'hour', 'dayofweek']
label = "count"
X, y = df[f], df[label]

all_result=[]
for c in combinations(f, 2) : 
    # print(df[list(c)])
    X_train, X_test, y_train, y_test = train_test_split(df[list(c)], df[label],test_size=0.3)
    model = RandomForestRegressor()
    model.fit(X_train, y_train)
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    #print("="*25, c, "="*25)
    #print("train : ", train_score)
    #print("test : ", test_score)
    sub_result = {"combination":str(c), 
                 "train":train_score,
                 "test":test_score}
    all_result.append(sub_result)
result_df = pd.DataFrame(all_result).sort_values(by="test")
result_df
```

```python
result_df.tail()
```

```python
fe = ['season', 'holiday', 'weather', 'temp', 'atemp', 'humidity', 'windspeed', 'year', 'month', 'day', 'dayofweek']
best_f=['workingday', 'hour']
```

```python
all_result=[]
for f in fe : 
    best_f.append(f)
    
    X_train, X_test, y_train, y_test = train_test_split(df[best_f], df[label],test_size=0.3)
    model = RandomForestRegressor()
    model.fit(X_train, y_train)
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    #print("="*25, c, "="*25)
    #print("train : ", train_score)
    #print("test : ", test_score)
    sub_result = {"combination":best_f.copy(), 
                 "train":train_score,
                 "test":test_score}
    all_result.append(sub_result)
    
    best_f.pop()
```

```python
result_df = pd.DataFrame(all_result).sort_values(by="test")
result_df.tail()
```

### 후진 선택
* 모든 컬럼을 넣고 하나씩 빼면서 확인하는 방식

```python
f = ['season', 'holiday', 'workingday', 'weather', 'temp',
       'atemp', 'humidity', 'windspeed', 
       'year', 'month', 'day', 'hour', 'dayofweek' ]
label = "count"

all_result = []

for item in combinations(f, len(f)-1):
    #print(item)
    target = list(item)
    X_train, X_test, y_train, y_test = \
            train_test_split(df[ target ] , df[label], test_size=0.3)
    model = RandomForestRegressor()
    model.fit( X_train, y_train )
    train_score = model.score( X_train, y_train )
    test_score = model.score( X_test, y_test )
    
    sub_result = {"combination" : set(f) - set(target), 
                  "train" : train_score ,
                 "test" : test_score }
    all_result.append( sub_result )
```

```python
result_df = pd.DataFrame(all_result).sort_values(by="test")
result_df.tail()
```

```python
f = ['season', 'holiday', 'workingday', 'weather', 'temp',
       'atemp', 'humidity', 'year', 'month', 'day', 'hour', 'dayofweek' ]
label = "count"

all_result = []
print( len(f) )
for item in combinations( f , len(f) - 1 ):
    #print( item )
    target = list(item)
    X_train, X_test, y_train, y_test = \
            train_test_split(df[ target ] , df[label], test_size=0.3)
    model = RandomForestRegressor()
    model.fit( X_train, y_train )
    train_score = model.score( X_train, y_train )
    test_score = model.score( X_test, y_test )
    
    sub_result = {"combination" : set(f) - set(target) , 
                  "train" : train_score ,
                 "test" : test_score }
    all_result.append( sub_result )
```

```python
result_df = pd.DataFrame(all_result).sort_values(by="test")
result_df.tail()
```

### RFE
* 사이킷런에서 제공하는 기능
* 변수의 중요도를 알려준다

```python
f = ['season', 'holiday', 'workingday', 'weather', 'temp',
       'atemp', 'humidity', 'windspeed', 
       'year', 'month', 'day', 'hour', 'dayofweek' ]
label = "count"
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[f] , df[label], test_size=0.3)
model = RandomForestRegressor()
model.fit( X_train, y_train )
train_score = model.score( X_train, y_train )
test_score = model.score( X_test, y_test )

print("train : ", train_score)
print("test : ", test_score)
```

```python
from sklearn.feature_selection import RFE
rfe = RFE(estimator=model)
rfe.fit(X_train, y_train)
print(rfe.ranking_)
X_train.columns
```

```python
df_rank = pd.DataFrame()
df_rank['ranking'] = rfe.ranking_
df_rank['features']= X_train.columns
df_rank.sort_values(by='ranking')
```

```python
f = [ 'workingday' ,'temp' , 'year' , 'month' ,'hour','dayofweek', 'humidity','atemp' ]
X_train, X_test, y_train, y_test = train_test_split(df[f] , df[label], test_size=0.3)

model = RandomForestRegressor()
model.fit(X_train, y_train)

print("train : ", model.score(X_train, y_train))
print("test : ", model.score(X_test, y_test))
```

### embed 방식
* tree 계열 알고리즘에만 존재한다

```python
f = ['season', 'holiday', 'workingday', 'weather', 'temp',
       'atemp', 'humidity', 'windspeed', 
       'year', 'month', 'day', 'hour', 'dayofweek' ]
label = "count"

X_train, X_test, y_train, y_test = train_test_split(df[f] , df[label], test_size=0.3)

model = RandomForestRegressor()
model.fit(X_train, y_train)

print("train : ", model.score(X_train, y_train))
print("test : ", model.score(X_test, y_test))
```

```python
model.feature_importances_
```

```python
df_im = pd.DataFrame()
df_im['rank'] = model.feature_importances_
df_im['columns'] = X_train.columns
df_im.sort_values(by="rank", ascending=False) #내림차순 ascending=False
```
